```python
import networkx as nx
import numpy as np
import pandas as pd
%matplotlib notebook
```

```python
# Instantiate the graph
G = nx.DiGraph()
# add node/edge pairs
G.add_edge("Austin", "Dallas", weight=200)
G.add_edge("Dallas", "Austin", weight=200)
G.add_edge("Austin", "Houston", weight=160)
G.add_edge("Dallas", "Denver", weight=780)
G.add_edge("Dallas", "Chicago", weight=900)
G.add_edge("Washington", "Dallas", weight=1300)
G.add_edge("Washington", "Atlanta", weight=600)
G.add_edge("Washington", "Dallas", weight=600)
G.add_edge("Denver", "Atlanta", weight=1400)
G.add_edge("Denver", "Chicago", weight=1000)
G.add_edge("Chicago", "Denver", weight=1000)
G.add_edge("Houston", "Atlanta", weight=800)
G.add_edge("Atlanta", "Houston", weight=800)
G.add_edge("Atlanta", "Washington", weight=600)
```

```python
nx.draw_networkx(G)
```

```python
print(nx.shortest_path(G,source="Austin",target="Chicago"))
```

```python
print(nx.shortest_path_length(G,source="Austin",target="Chicago", weight='weight'))
```

```python
for path in nx.all_simple_paths(G, source="Austin", target="Chicago"):
    print(path)
```

```python
print(nx.astar_path(G, "Austin","Chicago", weight='weight'))
```

```python
print(nx.astar_path_length(G, "Austin","Chicago", weight='weight'))
```

```python
print (nx.dijkstra_path(G, "Austin", "Chicago"))
```
