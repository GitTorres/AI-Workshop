```python
import networkx as nx
import numpy as np
import pandas as pd
%matplotlib notebook
```

```python
# Instantiate the graph
G1 = nx.Graph()
# add node/edge pairs
G1.add_edges_from([(0, 1),
                   (0, 2),
                   (0, 3),
                   (1, 7),
                   (1, 6),
                   (2, 4),
                   (2, 5),
                   (3, 8)])
```

```python
G1.nodes()
```

```python
G1.edges()
```

```python
# draw the network G1
nx.draw_networkx(G1)
```

```python
print(list(nx.dfs_edges(G1,0)))
```

```python
print(list(nx.bfs_edges(G1, 0)))
```
